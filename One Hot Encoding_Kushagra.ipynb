# One Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
data = {"Name" : ["Adam", "Dora", "Samuel", "Tina", "Mathew", "Sara", "Alice", "Frank", "Emma", "Peter"], 
        "Status" : ["Married", "Unmarried", "Divorced", "Unmarried", "Unmarried", "Married", "Married", "Divorced", "Married", "Divorced"], 
       "Education": ["Graduate", "Post Graduate", "Post Graduate", "Post Graduate", "Graduate", "Graduate", "Post Graduate", "Post Graduate", "Graduate", "Graduate"]}
```

```python
df = pd.DataFrame(data)
df
```

```python
status = pd.get_dummies(df["Status"])
status
```

```python
education = pd.get_dummies(df["Education"])
education
```

```python
new_df = pd.concat([df, status, education], axis = 1)
new_df
```

```python
new_df.drop(["Status", "Education"], axis = 1, inplace = True)
new_df
```
